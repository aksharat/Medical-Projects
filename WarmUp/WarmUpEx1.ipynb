```python
from PIL import Image
import matplotlib.pyplot as plt
import glob
```

```python

image_paths = glob.glob('/content/Practice_PNGandJPG/chestimage_JPG/*.jpg')

for imageName in image_paths:
    img = Image.open(imageName)
    print(imageName)
    plt.imshow(img,cmap='gray')
    plt.show()
```

```python

png_image_paths = glob.glob('/content/Practice_PNGandJPG/chestimage_PNG/*.png')

for pngimageName in png_image_paths:
    pngimg = Image.open(pngimageName)
    print(pngimageName)
    plt.imshow(pngimg,cmap='gray')
    plt.show()
```
